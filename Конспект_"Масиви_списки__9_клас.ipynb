# **Одновимірні масиви**




```python
#створення масиву
mas = [3,4,6,567,23,34]
print(mas)
print(mas[3], mas[5] )
mas = ["dfjdfj", "dghgh", "dhdfh"]
print(mas)
```

```python
#визначення довжини масиву (к-ть елементів у списку)
mas = [3,4,6,567,23,34]
print(len(mas))
```

```python
#змінення елемента списку
mas = ['a','b','c','d']
print(mas)
mas[2]='f'
print(mas)
```

```python
#додавання елемента в кінець списку
mas = ['a','b','c','d']
mas.append('e')
print(mas)
```

```python
#додавання в список елемента на вказану позицію
mas = ['a','b','c','d']
mas.insert(2,'r')
print(mas)
```

```python
#видалення елемента зі списку
mas = ['a','b','c','d']
del mas[2]
print(mas)
```

```python
#видалення конкретного значення зі списку
mas = ['a','b','c','d']
mas.remove('d')
print(mas)
```

# **Задачі**

Увести значення елементів списку з 5 чисел з клавіатури, піднести ці числа до квадрата та вивести отриманий список.

```python
#1 варіант розв'язку
mas = []
for i in range(5):
  x = float(input("Введіть елемент масиву "))
  mas.append(x)
print(mas)
for i in range(5):
  mas[i]=mas[i]**2
print(mas)
```

```python
#2 варіант розв'язку
a = list(map(float, input("введіть значення елементів списку через пробіл").split()))
print(a)
n = len(a)
for i in range(n):
  a[i]=a[i]**2
print(a)
a = list(map(str,a))
print(" ".join(a))
```

Задача1. Дано список [53, “Миколаїв”, “Херсон”, “Одеса”, 25].
Розробіть програму замінення числа 25 на число 37, додати новий елемент "Київ" виведення нового списку і визначення довжини списку.


```python
a = [53, "Миколаїв", "Херсон", "Одеса", 25]
a[4] = 37
a.append("Київ")
del a[0]
print(a)
print("Довжина списку", len(a))
```

Заповнити список з 32 елементів випадковими числами в інтервалі
[-85, 85], вивести його на екран; вивести на екран всі від’ємні числа зі створеного списку.

```python
from random import*
a = [i for i in sample(range(-85,85),32)]
#sample - метод із модуля random, який допомагає згенерувати
#задану к-ть елементів з певного діапазону
print(a)
n = len(a)
for i in range(n):
  if a[i]<0:
    print(a[i])
```
